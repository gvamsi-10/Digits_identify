## digits data classificatioin using svc

### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
datset = load_digits()
```

```python
dd = datset.data
dd
```

```python
datset.data[0]
```

```python
datset.target
```

```python
plt.cool
plt.matshow(datset.images[0])
```

```python
dir(datset)
```

```python
df = pd.DataFrame(dd, columns=[datset.feature_names])
```

```python
df['targt'] = datset.target
```

```python
df
```

```python
X = df.drop(['targt'],axis=1)
X
```

```python
y = df.targt
y
```

### splitting the Data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=38)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

### importing the required model

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
datset.target[67]
```

```python
svc.predict([datset.data[67]])
```

```python
svc.score(X_test,y_test)
```

### confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
sns.heatmap(data=cm, annot=True, cmap='Greens')
plt.xlabel('predicted')
plt.ylabel('true')
```

### using parameters

#### regularization(c)

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

#### gamma

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

```python
model_g = SVC(gamma=1)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

#### using rbf kernal

```python
rbf_model = SVC(kernel='rbf')
rbf_model.fit(X_train,y_train)
rbf_model.score(X_test,y_test)
```

```python
rbf_model = SVC(kernel='linear')
rbf_model.fit(X_train,y_train)
rbf_model.score(X_test,y_test)
```
